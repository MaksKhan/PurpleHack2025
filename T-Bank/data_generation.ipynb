```python
import numpy as np

# Создаем веса для распределения: 0% - 50%, остальные - по ~16.6%
weights = [5, 1, 1, 1]  # Соотношение 5:1:1:1
options = ['0%', '5%', '10%', '20%']

data['Персональный_кэшбэк'] = np.random.choice(
    options, 
    size=len(data),
    p=np.array(weights)/sum(weights)  # Нормализация весов
)
```

```python
import numpy as np

# Создаем распределение: 0 - 70%, 1-5 - по 6% каждый
weights = [8.0] + [0.4]*4  # 70% на 0 и 6% на каждое из чисел 1-5 (в сумме 30%)
values = [0, 1, 2, 3, 4]

data['Заказов_за_месяц'] = np.random.choice(
    values,
    size=len(data),
    p=np.array(weights)/sum(weights)  # Нормализуем веса до суммы=1
)

# Проверка распределения
print(data['Заказов_за_месяц'].value_counts(normalize=True).sort_index())
```

```python
import numpy as np

# Генерация цен, кратных 10, в диапазоне от 600 до 2000 рублей
data['Цена'] = np.random.choice(
    np.arange(600, 2001, 10),  # Генерируем значения от 600 до 2000 с шагом 10
    size=len(data)
)
data['Цена'] = data['Цена'].astype(str)
data['Цена'] += '₽'
# Проверка распределения
print(data['Цена'].head())
```

```python
data['Рейтинг'] = np.round(np.random.uniform(4.5, 5.0, size=len(data)), decimals=2)
data['Рейтинг'] = data['Рейтинг'].astype(str)
```

```python
data.head()
```

```python
# Генерация времени доставки: 20, 30, 40, 50 или 60 минут
delivery_times = [20, 30, 40, 50, 60]
data['Время_доставки'] = np.random.choice(
    delivery_times,
    size=len(data)
)
data['Время_доставки'] = data['Время_доставки'].astype(str) + ' мин'
```

```python
data.to_csv('../../Downloads/items_bd.csv', index=False) 
```
