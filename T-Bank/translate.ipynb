```python
model = ollama.Client()

def generate_response(prompt):
    response = model.generate(model='gemma2', prompt=prompt)
    return response['response']
```

```python
data = pd.read_csv('../../Downloads/final_bd.csv')
data.head()
```

```python
data.drop_duplicates(subset='hash', inplace=True)
```

```python
data['country_code'].unique()
```

```python
data.iloc[372]
```

```python
data = data[data['country_code'].isin(['UA', 'KZ', 'GE', 'KG', 'BG', 'NG', 'FR', 'IT'])]
```

```python
len(data)
```

```python
data['collection_section'].nunique()
```

```python
len(data)
```

```python
data['product_description'].iloc[20]
```

```python
product_names = []

for text in tqdm(data['product_name'].values):
    translated = generate_response(f"Переведи следующее предложение на русский язык. Выводи только переведенное предложение и больше ничего, предложение:\n{text}").strip()
    product_names.append(translated)
```

```python
len(product_names)
```

```python
data = data[:len(product_names)]
```

```python
data['product_name'] = product_names
```

```python
# data['collection_section'] = product_names
```

```python
data.to_csv('../../Downloads/final_bd.csv', index=False)
```
